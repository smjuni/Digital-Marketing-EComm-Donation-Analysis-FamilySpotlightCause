```python
import pandas as pd

pre_df = pd.read_excel("Dataset_for_cleaningV2.xlsx")

pre_df.head()
```

```python
print(pre_df.dtypes)
print(len(pre_df))
print(pre_df.count())
```

```python
#iterate thorugh and repeat the ord values by the number of times in basket count
value = [x for x in pre_df['orderID']]
# print(len(value))
repetitions = [y for y in pre_df['Basket Size']]
# print(repetitions) # Output: repetitions
z=[]
for i in range(len(value)):
  z.append((str(value[i]) + ' ') * repetitions[i]) # Output: (value[i] * repetitions[i])
pre_df['RepeatedOrdIDs'] = z

#turn Basket size back into a string to iterate through it
pre_df['Basket Size'] = pre_df['Basket Size'].astype(str)


# use if null order ids are present (there shouldn't be any)
nan_values_count = pd.isnull(pre_df).sum()
print(nan_values_count)
print(pre_df.count())
pre_df
```

```python

#make a new column with spaces replaced by commas in 'repeated ordvalues'
pre_df['Repeated OrdIDs2'] = pre_df['RepeatedOrdIDs'].apply(
    lambda s: (s.rstrip().replace(" ", ","))
)

pre_df = pre_df.drop(["RepeatedOrdIDs"], axis=1)
pre_df.columns

```

```python
#Drop columns that won't iterate (eg. have 1 value) 
pre_df = pre_df.drop(['orderID', 'Basket Size'], axis=1)
print(pre_df.dtypes)


```

```python
#  Check for NaN values
nan_values_count = pd.isnull(pre_df).sum()
print(nan_values_count)
print(pre_df.count())
```

```python
#explicitly cast Donation Amounts as string
pre_df['Donation Amounts'] = pre_df['Donation Amounts'].astype(str)
```

```python
# Split each column by comma into separate lists 
split_cols = [pre_df[col].str.split(',') for col in pre_df.columns]
print(len(split_cols)) #- 4
# print(split_cols[3]) #first column group


 #Check for NaN values
nan_values_count = pd.isnull(split_cols).sum()
print(nan_values_count)

# for QA & troubleshooting
# split_cols_df = pd.DataFrame(split_cols)
# split_cols_df.to_csv('split_cols_df.csv', index=False)
print(pre_df.count())
```

```python
#Split the columns into lists of lists (# of rows)
#tuples
rows = list(zip(*split_cols)) 

# QA
print(rows[0])
print(len(rows)) #rows

 
 #Check for NaN values
# nan_values_count = pd.isnull(split_cols).sum()
# print(nan_values_count)

#zip to get into each column group (# of columns:4)
rows2 = list(zip(*rows))
print(type(rows2)) #list of tuples
print(len(rows2))#- 4
print(rows2[0]) #first column group
```

```python
# create an empty list
blend = []
# Iterate through each row and zip the values together in the correct column
# # Note: rows2 is a list of tuples, where each tuple corresponds to a column in the original DataFrame
for i in range(len(rows)): #rows:
    zipped = []
    for r in rows2:
        val = r[i]
        if isinstance(val, (list, tuple)):
            zipped.append(val)
        else:
            zipped.append([val])
    blend.append(list(zip(*zipped)))
    
print(len(blend)) #(blend)
```

```python
# Convert the list of tuples into a DataFrame with specified columns
# Column order:
# Donation Category     100
# Donation Frequency    100
# Cause                 100
# Donation Amounts      100
# Repeated OrdIDs2      100

result_df = pd.DataFrame(columns=['Donation Category', 'Donation Frequency', 'Cause', 'Donation Amounts', 'Order ID'])
x=0
for i in range(len(blend)):
    for j in range(len(blend[i])):
            result_df.loc[x] = blend[i][j]
            x+=1

print(result_df.dtypes)
print(result_df.head())
```

```python
# # result_df.columns = result_df[col].str.lstrip() 
# for col in result_df.columns:
#     result_df.columns = result_df[col].str.lstrip() 
    
```

```python
# Create a unique ID for each group of OrdIDs
result_df['CounterID'] = result_df.groupby('Order ID').cumcount() + 1
result_df
```

```python
#Creating a unique key for each row
result_df['index_id'] = result_df.index.astype(str)


# create ID
result_df['unique_id']= (result_df['Order ID'] + "-" +(result_df['CounterID'].astype(str)) + "-" + result_df['index_id'])
#drop the helper columns
result_df.drop(['CounterID', 'index_id'], axis=1, inplace=True)

result_df


#QA that unique_id = number of rows
print(len(result_df["unique_id"].unique()))
print(len(result_df))
result_df
```

```python
 #Check for NaN values
nan_values_count = pd.isnull(result_df).sum()
print(nan_values_count)
print(len(result_df))
```
